```
race_results = spark.read.parquet('abfss://presentation@dlformulaone2024.dfs.core.windows.net/race_results')
```

```
demo_df = race_results.filter('race_year = 2020')
```

```
display(demo_df)
```

```
from pyspark.sql.functions import count, countDistinct, sum, groupBy
```

```
demo_df.select(count('*')).show()
```

```
demo_df.select(countDistinct('driver_nationality')).show()
```

```
demo_df.filter('driver_name = "Lewis Hamilton"').select(sum('points'), countDistinct('race_name')).show()
```

```
demo_df\
.groupBy('driver_name')\
.agg(sum('points'), countDistinct('race_name'))\
.show()
```
